```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd

# 크롬 드라이버 로드
driver = webdriver.Chrome()

# 각 리스트 초기화
title = []  # 상품 제목 리스트
url = []  # URL 리스트

driver.get('https://www.coupang.com/np/categories/486253?listSize=120&brand=&offerCondition=&filterType=rocket_luxury%2Crocket%2Ccoupang_global&isPriceRange=false&minPrice=&maxPrice=&page=1&channel=user&fromComponent=N&selectedPlpKeepFilter=&sorter=bestAsc&filter=&component=486153&rating=0&rocketAll=true')
time.sleep(3)  # 페이지가 완전히 로드될 때까지 대기

# 스크롤하여 모든 데이터를 로드
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # 스크롤 후 로딩 대기
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# 페이지 소스를 가져와서 BeautifulSoup으로 파싱
soup = bs(driver.page_source, 'html.parser')

# 모든 아이템을 가져옴
items = soup.find_all('li', class_='baby-product renew-badge')

# 각 아이템의 제목과 URL을 추출
for item in items:
    # 상품명
    try:
        a = item.find('div', class_='name').get_text(strip=True)
    except:
        a = 'null'
    title.append(a)

    # URL
    try:
        b = item.find('a', class_='baby-product-link').get('href')
    except:
        b = 'null'
    url.append(b)

# 드라이버 종료
driver.quit()

# 크롤링한 데이터를 데이터프레임으로 변환
raw_data = pd.DataFrame({
    'title': title,
    'url': url
})

# 결과 출력
print(raw_data)
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup as bs
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

# 크롬 드라이버 로드
driver = webdriver.Chrome(options=options)

driver.get('https://www.coupang.com/vp/products/7706148839?itemId=21726427953&vendorItemId=88775630832&sourceType=CATEGORY&categoryId=486153&isAddedCart=')
time.sleep(3)  # 페이지가 완전히 로드될 때까지 대기

# '상품평' 탭 클릭
try:
    review_tab = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, '#btfTab > ul.tab-titles > li:nth-child(2)'))
    )
    review_tab.click()
except Exception as e:
    print("Error clicking on the review tab:", e)
    driver.quit()
    raise

# 각 리스트 초기화
star = []  # 별점 리스트
date = []  # 작성 일자 리스트
review = []  # 리뷰 내용 리스트

# button:nth-child 번호를 4에서 시작하고 최대 10페이지까지 크롤링
for i in range(4, 14):  # 4에서 시작하여 13까지 반복 (최대 10페이지)
    time.sleep(3)
    
    # 페이지 소스를 가져와서 BeautifulSoup으로 파싱
    soup = bs(driver.page_source, 'html.parser')

    # 각 리뷰의 별점, 작성 일자, 리뷰 내용 추출
    items = soup.find_all('article', class_='sdp-review__article__list')
    for item in items:
        # 별점
        try:
            a = item.find('div', class_='sdp-review__article__list__info__product-info__star-orange')['data-rating']
        except (AttributeError, TypeError):
            a = 'null'
        star.append(a)

        # 작성 일자
        try:
            b = item.find('div', class_='sdp-review__article__list__info__product-info__reg-date').get_text(strip=True)
        except AttributeError:
            b = 'null'
        date.append(b)

        # 리뷰 내용
        try:
            c = item.find('div', class_='sdp-review__article__list__review__content').get_text(strip=True)
        except AttributeError:
            c = 'null'
        review.append(c)

    # 다음 리뷰 페이지로 이동
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, f'#btfTab > ul.tab-contents > li.product-review.tab-contents__content > div > div.sdp-review__article.js_reviewArticleContainer > section.js_reviewArticleListContainer > div.sdp-review__article__page.js_reviewArticlePagingContainer > button:nth-child({i})'))
        )
        next_button.click()
    except Exception as e:
        print("No more pages or error clicking on the next review button:", e)
        break

# 드라이버 종료
driver.quit()

# 크롤링한 데이터를 데이터프레임으로 변환
raw_data = pd.DataFrame({
    'star': star,
    'date': date,
    'review': review
})

# 결과 출력
print(raw_data)
```

## 문제점 해결 url을 통해서 이동하게 변경해줌

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup as bs
import pandas as pd

# WebDriver 초기화
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")


# 모든 리뷰를 저장할 빈 데이터프레임 생성
all_reviews = pd.DataFrame(columns=['title', 'star', 'date', 'review'])

for index, row in real_coopang_url.iterrows():
    title = row['title']
    url = row['url']
    
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    time.sleep(3)  # 페이지가 완전히 로드될 때까지 대기
    
    # '상품평' 탭 클릭
    try:
        review_tab = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, '#btfTab > ul.tab-titles > li:nth-child(2)'))
        )
        review_tab.click()
    except Exception as e:
        print(f"{title}의 리뷰 탭을 클릭하는 중 오류 발생: {e}")
        driver.quit()
        continue
    
    # 각 상품별 리스트 초기화
    star = []
    date = []
    review = []
    
    for i in range(4, 14):  # 4에서 시작하여 13까지 반복 (최대 10페이지)
        time.sleep(3)
        soup = bs(driver.page_source, 'html.parser')
        
        items = soup.find_all('article', class_='sdp-review__article__list')
        for item in items:
            try:
                a = item.find('div', class_='sdp-review__article__list__info__product-info__star-orange')['data-rating']
            except (AttributeError, TypeError):
                a = 'null'
            star.append(a)

            try:
                b = item.find('div', class_='sdp-review__article__list__info__product-info__reg-date').get_text(strip=True)
            except AttributeError:
                b = 'null'
            date.append(b)

            try:
                c = item.find('div', class_='sdp-review__article__list__review__content').get_text(strip=True)
            except AttributeError:
                c = 'null'
            review.append(c)
        
        # 다음 페이지 버튼 클릭
        try:
            next_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, f'#btfTab > ul.tab-contents > li.product-review.tab-contents__content > div > div.sdp-review__article.js_reviewArticleContainer > section.js_reviewArticleListContainer > div.sdp-review__article__page.js_reviewArticlePagingContainer > button:nth-child({i})'))
            )
            next_button.click()
        except Exception as e:
            print(f"{title}의 다음 리뷰 페이지 버튼을 클릭하는 중 오류 발생 또는 더 이상 페이지 없음: {e}")
            break
    
    driver.quit()
    
    # 현재 상품에 대한 데이터프레임 생성
    product_reviews = pd.DataFrame({
        'title': [title] * len(star),
        'star': star,
        'date': date,
        'review': review
    })
    
    # all_reviews 데이터프레임에 추가
    all_reviews = pd.concat([all_reviews, product_reviews], ignore_index=True)

# 필요시 CSV 또는 Excel로 저장
all_reviews.to_csv('coupang_reviews.csv', index=False)

# 결과 출력
print(all_reviews)
```
